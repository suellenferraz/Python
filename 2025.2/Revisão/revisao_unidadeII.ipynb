```python
# numeros aleatorios
import random
# basics
num = random.random()
print(num)

# simulando o jogo de dados
dados = random.randint(1, 6)
print(dados)
# basics
# simulando o jogo de dados
# supondo dois jogadores, em 10 rodadas do jogo de 1 dado
jogador1 = 0
jogador2 = 0
cont = 0
#for _ in range(10):
while cont < 10:
    cont += 1
    dado_jogador1 = random.randint(1, 6)
    dado_jogador2 = random.randint(1, 6)
    if dado_jogador1 > dado_jogador2:
        jogador1 += 1
    elif dado_jogador2 > dado_jogador1:
        jogador2 += 1
    else:
        jogador1 += 1
        jogador2 += 1
print('Jogador 1: ', jogador1)
print('Jogador 2: ', jogador2)
```

```python
# gerando uma lista de numeros aleatorios/notas entre 0 e 10
# com duas casas decimais
# calcularndo maior, menor e media sem e com funcoes
lista = []

for _ in range(5):
    num = round((random.random() * 10),2)
    lista.append(num)

print(lista)
print("Maior: ", max(lista))
print("Menor: ", min(lista))
print("Média:  {:.2f}".format((sum(lista)/len(lista))))
```

```python
# numa lista como saber se quem é int e quem é float
lista = ['1.1','2','3','4','5.5']
print(lista)

# transformar apenas os inteiros em int
lista = list(map(float, lista))
print(lista)
# primeira forma
for i in range(len(lista)):
    if lista[i] - int(lista[i]) == 0:
        lista[i] = int(lista[i])
    else: lista[i] = float(lista[i])
print(lista)

# segunda forma
lista = ['1.1','2','3','4','5.5']
for i in range(len(lista)):
    if "." in lista[i]:
        lista[i] = float(lista[i])
    else: lista[i] = int(lista[i])
print(lista)
```

```python
# Funções soma, maior, menor, aprovados_reprovados
lista_notas = [7.42, 1.1, 9.08, 1.94, 8]

def aprovados_recuperacao(lista):
    aprovados = []
    recuperacao = []
    for i in lista:
        if i >= 7:
            aprovados.append(i)
        else:
            recuperacao.append(i)
    return len(aprovados), len(recuperacao)

a, r = aprovados_recuperacao(lista_notas)
print(a)
print(r)
```

```python
# Dicionário de taxas de câmbio
taxas_cambio = {
    "USD_to_EUR": 0.85,  # 1 USD = 0.85 EUR
    "EUR_to_USD": 1 / 0.85  # 1 EUR = 1.18 USD (aproximadamente)
}
# Funções para converter moedas
def converter_moeda(valor, taxa_cambio):
    return valor * taxa_cambio
# Valores de exemplo para converter
valores_usd = [1, 50, 100, 500]
valores_eur = [1, 50, 100, 500]
# Converter USD para EUR
for i in range(len(valores_usd)):
    usd_para_eur = converter_moeda(valores_usd[i], taxas_cambio["USD_to_EUR"])
    valores_eur[i] = round(usd_para_eur,2)
# Converter EUR para USD
for i in range(len(valores_eur)):
    eur_para_usd = converter_moeda(valores_eur[i], taxas_cambio["EUR_to_USD"])
    valores_usd[i] = round(eur_para_usd,2)
# Imprimir os resultados
print("Valores em USD:", valores_usd)
print("Valores em EUR:", valores_eur)
```

```python
def calcular_bonificacao(medalhas, recorde_mundial, primeira_medalha_pais):
    # Dicionário com os valores das medalhas
    valores_medalhas = {
        "Ouro": 50000,
        "Prata": 30000,
        "Bronze": 10000
    }

    # Calcula a bonificação base somando os valores das medalhas
    bonificacao = sum(valores_medalhas[medalha] for medalha in medalhas)

    # Adiciona bônus por quebra de recorde mundial
    if recorde_mundial=='Sim':
        bonificacao += 100000

    # Adiciona bônus por primeira medalha do país na modalidade
    if primeira_medalha_pais == 'Sim':
        bonificacao += 50000

    return bonificacao

def leitura():
    medalhas = input().split(',')
    recorde_mundial = medalhas[-2]
    primeira_medalha_pais = medalhas[-1]
    medalhas = medalhas[:-2]
    return medalhas, recorde_mundial, primeira_medalha_pais

m,r,p = leitura()
print(calcular_bonificacao(m,r,p))
```

# Funções Recursivas

```python
# Funções recursivas
# vamos calcular a soma do intervalo
a = 10; b = 12

def soma_normal(a,b):
    soma = 0
    for i in range(a, b+1):
        soma += i
    return soma

print(soma_normal(a, b))

def soma_rec(a, b):
    if a > b: return 0
    else: return a + soma_rec(a+1, b)
def soma_rec2(a, b):
    if a == b: return b
    else: return b + soma_rec2(a, b-1)
def soma_rec3(a, b):
    if a == b: return a
    else: return a + soma_rec3(a+1, b)
def fatorial(n):
    if n == 0 or n==1: return 1
    else: return n * fatorial(n-1)

print(soma_rec3(a, b))
print(soma_rec2(a, b))
print(soma_rec(a, b))
```
